```python
from typing import List
```

```python
def transform_clause(clause: List[str], cnf3: List[List[str]], c_index: int):
    k = len(clause)
    z_var = [f"z_{c_index}{i+1}" for i in range(k - 3)]
    cnf.append([clause[0], clause[1], z_var[0]])
    for i in range(1, k-3):
        cnf3.append([f"~{z_var[i-1]}", clause[i+1], z_var[i]])
    cnf3.append([f"~{z_var[k-4]}", clause[k-2], clause[k-1]])
    return cnf3
```

```python
def cnf_to_3_cnf(cnf_form, cnf3 = []):
    """
    This function turn CNF_FORM of a formula to 3CNF_FORM.
    If the given cnf_form is empty => return []
    Else we analyse each clause.
    if the len of the current clause is great then 3 we turn it into a cube of clauses using combination of originals clauses and fresh variables
    if not we keep the original clause as it's.
    At the end we return the 3CNF_FORM
    """
    if len(cnf_form) == 0:
        return
    for clause in cnf_form:
        if len(clause) <= 3:
            cnf3.append(clause)
        else:
            c_index = cnf_form.index(clause)
            cnf3 = transform_clause(clause, cnf3, c_index)
    return cnf3
    
```

```python
cnf = [['D', 'H', 'S', 'k', 'L'], ['R', 'O', 'X', 'Y', 'A', 'B', 'C']]
cnf_to_3_cnf(cnf)
```
